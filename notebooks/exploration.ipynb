## Dataset Description

This dataset contains credit card transactions made by European cardholders.

Features:

- Time  seconds elapsed between transactions
- Amount → transaction amount
- Class → target variable

Class values:
- 0 = normal transaction
- 1 = fraud

```python
# Load dataset containing credit card transactions
# The dataset includes both legitimate and fraudulent transactions

df = pd.read_csv("../data/creditcard.csv")

# Display first rows of dataset
df.head()
```

## Class Distribution

The dataset is highly imbalanced.

Fraudulent transactions represent a very small percentage
of the total number of transactions.

This is an important challenge for machine learning models.

```python
plt.figure(figsize=(10,5))
sns.histplot(df["Amount"], bins=50)
plt.title("Transaction Amount Distribution")
plt.show()
```

## Transaction Amount Distribution

Most transactions involve small amounts.

However, some transactions show larger values,
which may influence fraud detection.

```python
corr = df.corr()

plt.figure(figsize=(12,10))
sns.heatmap(corr, cmap="coolwarm")
plt.title("Feature Correlation")
plt.show()
```

## Feature Correlation

The correlation matrix helps us understand relationships
between variables in the dataset.

Highly correlated features may influence machine learning models.

## EDA Conclusion

Key observations:

- The dataset has no missing values
- The dataset is highly imbalanced
- Fraud cases represent a very small fraction of the data
- Transaction amounts vary significantly
